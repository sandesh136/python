```python
import pandas as pd
final=pd.DataFrame()
url = 'https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;spanmax1=31+Dec+2020;spanmin1=01+Jan+1965;spanval1=span;template=results;type=aggregate;view=results'

df_list1 = pd.read_html(url)

for i in range(2,25):
    url2='https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;page={};spanmax1=31+Dec+2020;spanmin1=01+Jan+1965;spanval1=span;template=results;type=aggregate;view=results'
    url1=url2.format(i)
    df_list = pd.read_html(url1)
    df=df_list[2]
    final=final.append(df)
final_df = df_list1[2]
data = final_df.append(final)
data = data.reset_index()
data = data.drop(['index','BR','Unnamed: 4','Unnamed: 8'],axis=1)
```

```python
data=data[data['Winner'].notna()]
```

```python
data[['First Team','Second Team']]=data['Match'].str.split(' v ',1,expand=True)
data=data.drop(['Match'],axis=1)
data
```
